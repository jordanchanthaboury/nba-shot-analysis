```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('nba_team_stats_24_25_121724.csv')

# Display the first few rows of the DataFrame to confirm it loaded correctly
df.head()
```

```python
# Select specific columns
selected_columns = ['Team', 'G', 'PTS', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'FT', 'FTA', 'FT%']

# Create a new DataFrame with only the selected columns
shot_focused_df = df[selected_columns]

# Display the first few rows of the new DataFrame
shot_focused_df.head(15)
```

```python
# Calculate the expected value of 2-point shots using .loc to avoid the warning
shot_focused_df.loc[:, 'Expected 2P Value'] = 2 * (shot_focused_df['2P%'])

# Display the updated DataFrame
shot_focused_df.head(15)
```

```python
# Calculate the expected value of 3-point shots using .loc to avoid the warning
shot_focused_df.loc[:, 'Expected 3P Value'] = 3 * (shot_focused_df['3P%'])

# Display the updated DataFrame
shot_focused_df.head(30)
```

```python
# Calculate the combined expected value of 2P and 3P
shot_focused_df.loc[:, 'Combined Expected Value'] = shot_focused_df['Expected 2P Value'] + shot_focused_df['Expected 3P Value']

# Calculate the shot expected value of 2P and 3P
shot_focused_df.loc[:, 'Shot Expected Value'] = (((shot_focused_df['Expected 2P Value'] )
                                                      * (shot_focused_df['2PA'] / shot_focused_df['FGA'])) 
                                                     + (shot_focused_df['Expected 3P Value']
                                                        * (shot_focused_df['3PA'] / shot_focused_df['FGA'])))

# Display the updated DataFrame
shot_focused_df.head(30)
```

```python
# Create a new column to compare the expected values
shot_focused_df['Greater EV'] = shot_focused_df.apply(
    lambda row: '2' if row['Expected 2P Value'] > row['Expected 3P Value'] else ('3' if row['Expected 2P Value'] < row['Expected 3P Value'] else 'Equal EV'),
    axis=1
)

# Display the updated DataFrame with the new column
shot_focused_df.head(15)
```

```python
# Sort the DataFrame by 'Combined Expected Value' in descending order
shot_focused_df_sorted = shot_focused_df.sort_values(by='Combined Expected Value', ascending=False)

# Display the sorted DataFrame
shot_focused_df_sorted.head(30)
```

```python
# Calculate the ideal number of 2-point and 3-point attempts based on their expected proportions
shot_focused_df['2P Expected Proportion'] = shot_focused_df['Expected 2P Value'] / shot_focused_df['Combined Expected Value']
shot_focused_df['3P Expected Proportion'] = shot_focused_df['Expected 3P Value'] / shot_focused_df['Combined Expected Value']

# Calculate the ideal 2P and 3P attempts using FGA (Total Shots)
shot_focused_df['Ideal 2P Attempts'] = shot_focused_df['2P Expected Proportion'] * shot_focused_df['FGA']
shot_focused_df['Ideal 3P Attempts'] = shot_focused_df['3P Expected Proportion'] * shot_focused_df['FGA']

# Calculate the ideal 2P and 3P makes using FG (Total Shots)
shot_focused_df['Ideal 2P Makes'] = shot_focused_df['2P Expected Proportion'] * shot_focused_df['FG']
shot_focused_df['Ideal 3P Makes'] = shot_focused_df['3P Expected Proportion'] * shot_focused_df['FG']

# Calculate the ideal 2P and 3P percentage
shot_focused_df['Ideal 2P%'] = shot_focused_df['Ideal 2P Makes'] / shot_focused_df['Ideal 2P Attempts']
shot_focused_df['Ideal 3P%'] = shot_focused_df['Ideal 3P Makes'] / shot_focused_df['Ideal 3P Attempts']

# Display the updated DataFrame with ideal attempts
shot_focused_df[['Team', 'FG', 'FGA', 
                 '2P Expected Proportion', '2P Expected Proportion',
                 '2P', '2PA','2P%', '3P','3PA','3P%',
                 'Ideal 2P Attempts', 'Ideal 3P Attempts',
                 'Ideal 2P Makes', 'Ideal 3P Makes',
                 'Ideal 2P%', 'Ideal 3P%']].head(30)
```

```python
# Calculate the net difference for 2PA and Ideal 2P Attempts
shot_focused_df['2P Attempt Difference'] = shot_focused_df['Ideal 2P Attempts'] - shot_focused_df['2PA']

# Calculate the net difference for 3PA and Ideal 3P Attempts
shot_focused_df['3P Attempt Difference'] = shot_focused_df['Ideal 3P Attempts'] - shot_focused_df['3PA']

# Display the updated DataFrame with the new columns
shot_focused_df[['Team', 'FGA', '2PA', 'Ideal 2P Attempts', '2P Attempt Difference', '3PA', 'Ideal 3P Attempts', '3P Attempt Difference']].head(30)
```

```python
# Calculate the total points from Ideal 3PA
shot_focused_df.loc[:, 'Ideal 2P Total'] = shot_focused_df['Ideal 2P Makes'] * 2

# Calculate the total points from Ideal 3PA
shot_focused_df.loc[:, 'Ideal 3P Total'] = shot_focused_df['Ideal 3P Makes'] * 3

# EV total PPG
shot_focused_df.loc[:, 'Ideal Total'] = shot_focused_df['Ideal 3P Total'] + shot_focused_df['Ideal 2P Total']

# EV total PPG w/ FT
shot_focused_df.loc[:, 'Ideal Total w FTM'] = shot_focused_df['Ideal 3P Total'] + shot_focused_df['Ideal 2P Total'] + shot_focused_df['FT']

shot_focused_df
```

```python
# Calculate the total points difference from actual - EV PPG
shot_focused_df.loc[:, 'Actual vs. EVPPG'] = shot_focused_df['Ideal Total w FTM'] - shot_focused_df['PTS']

shot_focused_df[['Team','PTS', '2P', '2PA', '2P%', '3P', '3PA', '3P%',
                 'Ideal 2P Makes', 'Ideal 2P Attempts','2P Attempt Difference',
                 'Ideal 3P Makes','Ideal 3P Attempts', '3P Attempt Difference',
                 'Ideal Total w FTM', 'Greater EV', 'Actual vs. EVPPG']]
```
