```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('nba_team_stats_shot_zones.csv')

# Display the first few rows of the DataFrame to confirm it loaded correctly
df.head()
```

```python
# Load the CSV files into DataFrames
df1 = pd.read_csv('nba_team_stats_shot_zones.csv')
df2 = pd.read_csv('nba_team_stats.csv')

# Perform the join on the 'Team' column
merged_df = pd.merge(df1, df2, on='Team', how='inner')  # You can adjust the 'how' parameter as needed

# Display the first few rows of the merged DataFrame
merged_df.head(5)
```

```python
# Select specific columns
selected_columns = ['Team','RA_FGM', 'RA_FGA', 'RA_FG%',
                    'NRA_FGM', 'NRA_FGA', 'NRA_FG%',
                    'MR_FGM', 'MR_FGA', 'MR_FG%',
                    'LC3_FGM', 'LC3_FGA', 'LC3_FG%',
                    'RC3_FGM', 'RC3_FGA', 'RC3_FG%',
                    'AB3_FGM', 'AB3_FGA', 'AB3_FG%',]

# Create a new DataFrame with only the selected columns
shot_zone_focused_df = df[selected_columns]

# Display the first few rows of the new DataFrame
shot_zone_focused_df.head(5)
```

```python
# Calculate Total_FGA
shot_zone_focused_df['Total_FGA'] = (shot_zone_focused_df['RA_FGA'] 
                                     + shot_zone_focused_df['NRA_FGA']
                                     + shot_zone_focused_df['MR_FGA']
                                     + shot_zone_focused_df['LC3_FGA']
                                     + shot_zone_focused_df['RC3_FGA']
                                     + shot_zone_focused_df['AB3_FGA'])

# Calculate Total_FGM
shot_zone_focused_df['Total_FGM'] = (shot_zone_focused_df['RA_FGM'] 
                                     + shot_zone_focused_df['NRA_FGM']
                                     + shot_zone_focused_df['MR_FGM']
                                     + shot_zone_focused_df['LC3_FGM']
                                     + shot_zone_focused_df['RC3_FGM']
                                     + shot_zone_focused_df['AB3_FGM'])

# Calculate total_points
shot_zone_focused_df['Total_Points'] = ((shot_zone_focused_df['RA_FGM'] * 2)
                                        + (shot_zone_focused_df['NRA_FGM'] * 2)
                                        + (shot_zone_focused_df['MR_FGM'] * 2)
                                        + (shot_zone_focused_df['LC3_FGM'] * 3)
                                        + (shot_zone_focused_df['RC3_FGM'] * 3)
                                        + (shot_zone_focused_df['AB3_FGM'] * 3))

# ---

# Proportion of shots taken from RA
shot_zone_focused_df['RA_PA'] = (shot_zone_focused_df['RA_FGA'] / shot_zone_focused_df['Total_FGA'])

# Proportion of shots taken from NRA
shot_zone_focused_df['NRA_PA'] = (shot_zone_focused_df['NRA_FGA'] / shot_zone_focused_df['Total_FGA'])

# Proportion of shots taken from MR
shot_zone_focused_df['MR_PA'] = (shot_zone_focused_df['MR_FGA'] / shot_zone_focused_df['Total_FGA'])

# Proportion of shots taken from LC3
shot_zone_focused_df['LC3_PA'] = (shot_zone_focused_df['LC3_FGA'] / shot_zone_focused_df['Total_FGA'])

# Proportion of shots taken from RC3
shot_zone_focused_df['RC3_PA'] = (shot_zone_focused_df['RC3_FGA'] / shot_zone_focused_df['Total_FGA'])

# Proportion of shots taken from AB3
shot_zone_focused_df['AB3_PA'] = (shot_zone_focused_df['AB3_FGA'] / shot_zone_focused_df['Total_FGA'])

# ---

# Calculate the expected value of restricted area shots
shot_zone_focused_df['EV_RA'] = 2 * (shot_zone_focused_df['RA_FG%'] / 100)

# Calculate the expected value of non-restricted area shots
shot_zone_focused_df['EV_NRA'] = 2 * (shot_zone_focused_df['NRA_FG%'] / 100)

# Calculate the expected value of shots in the paint
shot_zone_focused_df['EV_Paint'] = ((shot_zone_focused_df['EV_NRA'] + shot_zone_focused_df['EV_RA']) / 2)

# --

# Calculate the expected value of mid-range shots
shot_zone_focused_df['EV_MR'] = 2 * (shot_zone_focused_df['MR_FG%'] / 100)

# --

# Calculate the expected value of left corner 3 shots
shot_zone_focused_df['EV_LC3'] = 3 * (shot_zone_focused_df['LC3_FG%'] / 100)

# Calculate the expected value of right corner 3 shots
shot_zone_focused_df['EV_RC3'] = 3 * (shot_zone_focused_df['RC3_FG%'] / 100)

# Calculate the expected value of above break 3 shots
shot_zone_focused_df['EV_AB3'] = 3 * (shot_zone_focused_df['AB3_FG%'] / 100)

# Calculate the expected value of threes
shot_zone_focused_df['EV_3'] = ((shot_zone_focused_df['EV_LC3'] 
                                 + shot_zone_focused_df['EV_RC3'] 
                                 + shot_zone_focused_df['EV_AB3']) / 3)

#--

shot_zone_focused_df[['Team',
                      'EV_RA','EV_NRA', 'EV_Paint',
                      'EV_MR', 
                      'EV_LC3','EV_RC3','EV_AB3','EV_3']]
```

```python
shot_zone_focused_df['EV Total'] = (
    (shot_zone_focused_df['EV_RA']) +
    shot_zone_focused_df['EV_NRA'] +
    shot_zone_focused_df['EV_MR'] +
    shot_zone_focused_df['EV_LC3'] +
    shot_zone_focused_df['EV_RC3'] +
    shot_zone_focused_df['EV_AB3']
)

shot_zone_focused_df[['Team','EV Total']]
```

```python
# Calculate how many of each type of shot the teams should take given their total EV: RA
# Calculate the ideal proportiong of RA, the number attempts, the number of makes based on their expected proportions
shot_zone_focused_df['EV_RA_Proportion'] = shot_zone_focused_df['EV_RA'] / shot_zone_focused_df['EV Total']

shot_zone_focused_df['EV_RA_Attempts'] = shot_zone_focused_df['EV_RA_Proportion'] * shot_zone_focused_df['Total_FGA']
shot_zone_focused_df['EV_RA_Makes'] = shot_zone_focused_df['EV_RA_Proportion'] * shot_zone_focused_df['Total_FGM']

# This is going to read we want to take less or more shots 
shot_zone_focused_df['EV_RA_Attempts_Diff'] = shot_zone_focused_df['EV_RA_Attempts'] - shot_zone_focused_df['RA_FGA']
shot_zone_focused_df['EV_RA_Makes_Diff'] = shot_zone_focused_df['EV_RA_Makes'] - shot_zone_focused_df['RA_FGM']

shot_zone_focused_df[['Team'
                      ,'RA_PA','EV_RA_Proportion'
                      ,'RA_FGM','RA_FGA'
                      ,'EV_RA_Makes','EV_RA_Attempts'
                      ,'EV_RA_Makes_Diff', 'EV_RA_Attempts_Diff']]
```

```python
# Calculate how many of each type of shot the teams should take given their total EV: NRA
# Calculate the ideal proportion of NRA, the number attempts, the number of makes based on their expected proportions
shot_zone_focused_df['EV_NRA_Proportion'] = shot_zone_focused_df['EV_NRA'] / shot_zone_focused_df['EV Total']

shot_zone_focused_df['EV_NRA_Attempts'] = shot_zone_focused_df['EV_NRA_Proportion'] * shot_zone_focused_df['Total_FGA']
shot_zone_focused_df['EV_NRA_Makes'] = shot_zone_focused_df['EV_NRA_Proportion'] * shot_zone_focused_df['Total_FGM']

# This is going to read we want to take less or more shots 
shot_zone_focused_df['EV_NRA_Attempts_Diff'] = shot_zone_focused_df['EV_NRA_Attempts'] - shot_zone_focused_df['NRA_FGA']
shot_zone_focused_df['EV_NRA_Makes_Diff'] = shot_zone_focused_df['EV_NRA_Makes'] - shot_zone_focused_df['NRA_FGM']

shot_zone_focused_df[['Team'
                      ,'NRA_PA','EV_NRA_Proportion'
                      ,'NRA_FGM','NRA_FGA'
                      ,'EV_NRA_Makes','EV_NRA_Attempts'
                      ,'EV_NRA_Makes_Diff', 'EV_NRA_Attempts_Diff']]
```
